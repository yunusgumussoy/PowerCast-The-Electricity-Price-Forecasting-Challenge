```python
from catboost import CatBoostRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np


# Split data chronologically (80% training, 20% testing)
train_size = int(0.8 * len(numeric_data))
train_data = numeric_data[:train_size]
test_data = numeric_data[train_size:]

# Prepare features and target
features_train = train_data.drop(columns=['Day-ahead_prices_Germany/Luxembourg [€/MWh] Original resolutions'])
target_train = train_data['Day-ahead_prices_Germany/Luxembourg [€/MWh] Original resolutions']

features_test = test_data.drop(columns=['Day-ahead_prices_Germany/Luxembourg [€/MWh] Original resolutions'])
target_test = test_data['Day-ahead_prices_Germany/Luxembourg [€/MWh] Original resolutions']
```

```python

# CatBoost

# Initialize CatBoostRegressor
cat_model = CatBoostRegressor(iterations=500, depth=6, learning_rate=0.1, silent=True)

# Train the model
cat_model.fit(features_train, target_train)

# Make predictions
predictions = cat_model.predict(features_test)

# Calculate evaluation metrics
mse = mean_squared_error(target_test, predictions)
rmse = np.sqrt(mse)
mae = mean_absolute_error(target_test, predictions)

# Print metrics
print(f'CatBoost - MSE: {mse}')
print(f'CatBoost - RMSE: {rmse}')
print(f'CatBoost - MAE: {mae}')
```

```python

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

# Initialize RandomForestRegressor
rf_model = RandomForestRegressor(n_estimators=500, max_depth=6, random_state=42, n_jobs=-1)

# Train the model
rf_model.fit(features_train, target_train)

# Make predictions
rf_predictions = rf_model.predict(features_test)

# Calculate evaluation metrics
rf_mse = mean_squared_error(target_test, rf_predictions)
rf_rmse = np.sqrt(rf_mse)
rf_mae = mean_absolute_error(target_test, rf_predictions)

# Print metrics
print(f'Random Forest - MSE: {rf_mse}')
print(f'Random Forest - RMSE: {rf_rmse}')
print(f'Random Forest - MAE: {rf_mae}')
```

```python

from xgboost import XGBRegressor

# Rename columns to remove special characters
features_train_clean = features_train.rename(columns=lambda x: x.replace('[', '').replace(']', '').replace('<', '').replace(' ', '_'))
features_test_clean = features_test.rename(columns=lambda x: x.replace('[', '').replace(']', '').replace('<', '').replace(' ', '_'))

# Initialize XGBoost Regressor
xgb_model = XGBRegressor(n_estimators=500, max_depth=6, learning_rate=0.1, objective='reg:squarederror', n_jobs=-1, random_state=42)

# Train the model with cleaned feature names
xgb_model.fit(features_train_clean, target_train)

# Make predictions
xgb_predictions = xgb_model.predict(features_test_clean)

# Calculate evaluation metrics
xgb_mse = mean_squared_error(target_test, xgb_predictions)
xgb_rmse = np.sqrt(xgb_mse)
xgb_mae = mean_absolute_error(target_test, xgb_predictions)

# Print metrics
print(f'XGBoost - MSE: {xgb_mse}')
print(f'XGBoost - RMSE: {xgb_rmse}')
print(f'XGBoost - MAE: {xgb_mae}')
```

```python
# What percentage of predictions are correct when classified as rising, falling, or stable?
```

```python
# Make sure to use the same feature set as during model training
features = numeric_data.drop(columns=['Day-ahead_prices_Germany/Luxembourg [€/MWh] Original resolutions'])
predictions = model.predict(features)

# Shift actual and predicted prices to calculate the direction of change
numeric_data['Actual_Direction'] = numeric_data['Day-ahead_prices_Germany/Luxembourg [€/MWh] Original resolutions'].diff().shift(-1)
numeric_data['Predicted_Direction'] = pd.Series(predictions).diff().shift(-1)

# Classify as rising (1), falling (-1), or stable (0)
numeric_data['Actual_Direction_Label'] = numeric_data['Actual_Direction'].apply(lambda x: 1 if x > 0 else (-1 if x < 0 else 0))
numeric_data['Predicted_Direction_Label'] = numeric_data['Predicted_Direction'].apply(lambda x: 1 if x > 0 else (-1 if x < 0 else 0))
```

```python
# Compare predicted direction with actual direction
correct_directions = (numeric_data['Actual_Direction_Label'] == numeric_data['Predicted_Direction_Label'])

# Calculate Directional Accuracy for each category (rising, falling, stable)
total_predictions = len(correct_directions)
correct_rising = len(correct_directions[correct_directions & (numeric_data['Actual_Direction_Label'] == 1)])
correct_falling = len(correct_directions[correct_directions & (numeric_data['Actual_Direction_Label'] == -1)])
correct_stable = len(correct_directions[correct_directions & (numeric_data['Actual_Direction_Label'] == 0)])

directional_accuracy = {
    "Rising Accuracy (%)": (correct_rising / total_predictions) * 100,
    "Falling Accuracy (%)": (correct_falling / total_predictions) * 100,
    "Stable Accuracy (%)": (correct_stable / total_predictions) * 100
}

print(directional_accuracy)
```

```python
# Compute actual and predicted volatility using absolute price differences
numeric_data['Actual_Volatility'] = numeric_data['Day-ahead_prices_Germany/Luxembourg [€/MWh] Original resolutions'].diff().abs()
numeric_data['Predicted_Volatility'] = pd.Series(predictions).diff().abs()

# Compute rolling standard deviation as an alternative volatility measure (e.g., 7-day window)
numeric_data['Actual_Rolling_Volatility'] = numeric_data['Actual_Volatility'].rolling(window=7).std()
numeric_data['Predicted_Rolling_Volatility'] = numeric_data['Predicted_Volatility'].rolling(window=7).std()
```

```python
# Extreme Price Movement Detection
```

```python
# Can the model correctly predict sharp price spikes (>15% increase or decrease)?
```

```python
# Compute percentage price changes for actual and predicted values
numeric_data['Actual_Percentage_Change'] = numeric_data['Day-ahead_prices_Germany/Luxembourg [€/MWh] Original resolutions'].pct_change() * 100
numeric_data['Predicted_Percentage_Change'] = pd.Series(predictions).pct_change() * 100

# Define extreme movement threshold (absolute percentage change > 15%)
threshold = 15

# Identify extreme price movements
numeric_data['Actual_Spike'] = numeric_data['Actual_Percentage_Change'].abs() > threshold
numeric_data['Predicted_Spike'] = numeric_data['Predicted_Percentage_Change'].abs() > threshold

# Count detected spikes
actual_spikes = numeric_data['Actual_Spike'].sum()
predicted_spikes = numeric_data['Predicted_Spike'].sum()
correctly_predicted_spikes = (numeric_data['Actual_Spike'] & numeric_data['Predicted_Spike']).sum()

print(f"Total Actual Spikes: {actual_spikes}")
print(f"Total Predicted Spikes: {predicted_spikes}")
print(f"Correctly Predicted Spikes: {correctly_predicted_spikes}")
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Convert boolean values to 0/1 for classification metrics
y_true = numeric_data['Actual_Spike'].astype(int)
y_pred = numeric_data['Predicted_Spike'].astype(int)

precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)

print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(numeric_data.index, numeric_data['Actual_Percentage_Change'], label="Actual % Change", color='blue', alpha=0.7)
plt.plot(numeric_data.index, numeric_data['Predicted_Percentage_Change'], label="Predicted % Change", color='red', alpha=0.7)

# Highlight extreme price spikes
plt.scatter(numeric_data.index[numeric_data['Actual_Spike']], 
            numeric_data['Actual_Percentage_Change'][numeric_data['Actual_Spike']], 
            color='blue', label="Actual Spike", marker='o', edgecolor='black', s=100)

plt.scatter(numeric_data.index[numeric_data['Predicted_Spike']], 
            numeric_data['Predicted_Percentage_Change'][numeric_data['Predicted_Spike']], 
            color='red', label="Predicted Spike", marker='x', s=100)

plt.xlabel("Time")
plt.ylabel("Percentage Change")
plt.title("Extreme Price Movement Detection: Actual vs. Predicted")
plt.legend()
plt.show()
```

```python
# Separate upward and downward spikes
numeric_data['Actual_Spike_Up'] = numeric_data['Actual_Percentage_Change'] > threshold
numeric_data['Actual_Spike_Down'] = numeric_data['Actual_Percentage_Change'] < -threshold

numeric_data['Predicted_Spike_Up'] = numeric_data['Predicted_Percentage_Change'] > threshold
numeric_data['Predicted_Spike_Down'] = numeric_data['Predicted_Percentage_Change'] < -threshold

# Compute precision & recall separately
precision_up = precision_score(numeric_data['Actual_Spike_Up'], numeric_data['Predicted_Spike_Up'])
recall_up = recall_score(numeric_data['Actual_Spike_Up'], numeric_data['Predicted_Spike_Up'])

precision_down = precision_score(numeric_data['Actual_Spike_Down'], numeric_data['Predicted_Spike_Down'])
recall_down = recall_score(numeric_data['Actual_Spike_Down'], numeric_data['Predicted_Spike_Down'])

print(f"Upward Spike Precision: {precision_up:.4f}, Recall: {recall_up:.4f}")
print(f"Downward Spike Precision: {precision_down:.4f}, Recall: {recall_down:.4f}")
```

```python
# Confidence Intervals & Probability Forecasting
```

```python
# Calculate residuals (differences between actual and predicted prices)
train_predictions = model.predict(features_train)
residuals = target_train - train_predictions
```

```python
# Calculate the standard deviation of residuals
residual_std = np.std(residuals)
```

```python
# Predict on the test set
test_predictions = model.predict(features_test)

# Calculate the prediction interval (95% confidence)
lower_bound = test_predictions - 1.96 * residual_std
upper_bound = test_predictions + 1.96 * residual_std
```

```python
# Plot actual, predicted prices, and the 95% confidence interval
plt.figure(figsize=(14, 8))

# Plot actual prices (black dotted line)
plt.plot(target_test.values, label="Actual Prices", color='black', linestyle='dotted', alpha=0.8)

# Plot predicted prices (blue line)
plt.plot(test_predictions, label="Predicted Prices", color='orange', linewidth=2)

# Plot the 95% prediction interval (with a contrasting color)
plt.fill_between(range(len(lower_bound)), lower_bound, upper_bound, color='blue', alpha=0.9, label="95% Prediction Interval")

# Adding labels and title
plt.xlabel("Time", fontsize=12)
plt.ylabel("Day-Ahead Price [€/MWh]", fontsize=12)
plt.title("Predicted Price with 95% Confidence Interval", fontsize=14)

# Add gridlines for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Add legend
plt.legend(loc='upper left', fontsize=12)

# Tight layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Get the minimum and maximum predicted intervals
min_lower_bound = np.min(lower_bound)
max_upper_bound = np.max(upper_bound)

print(f"95% Prediction Interval: [{min_lower_bound:.2f}, {max_upper_bound:.2f}] €/MWh")
```
